## DEcorator

```python

def ad(a,b):
    print(a/b)
    
ad(2,5)
```

```python
def dec(fun):
    def inner(a,b):
        if a>b:
            return fun(a,b) 
    return inner

ddc=dec(ad)
ddc(6,5)
```
