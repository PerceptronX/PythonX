## Python - Control Flow and conditional Statement

## for Loop

```python
for i,j in zip(range(0,5),range(10,15)):
    print(i,j)
    
```

```python
x=[10,20,30,40,50]
for i in x:
    print(i)
```

```python
lt=[10,20,30,40,50]
l=len(lt)
for i in range(0,l):
    print(lt[i])
    print("i value",i)
```

```python
words=["absdb","sbdbf","dbndjfj","jfjfj","weere","sdcx"]
for i in words:
    print(i)
```

```python
for i in range(0,10,2):
    print(i)
```

## WHILE LOOP

```python
vl=[1,3,2,4,7,5]
v=len(vl)-1
while(v>=0):
    print(vl[v])
    v=v-1
```

## if  else condition

```python
"""
if (condition):
    statement
else:
    statement 
    
"""
```

```python
if 2>5:
    print("statement is true")
else:
    print("statement is false")
```

```python
## find value in the list which are greater than 5
lst=[1,4,2,6,15,20]
for i in lst:
    if i>5:
        print("greater number than 5 is",i)
```

## break statement

```python
for i in range(0,20):
    if i==10:
        break
    else :
        print("values====",i)
```

## continious statement

```python
for i in range(0,10):
    if i>5:
        print("i val==",i)
        continue 
    else:
        print(i)
```
